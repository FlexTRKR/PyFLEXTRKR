### Rainfall volume
Volume = rain_depth $\times$ area = $mm$ $h^{-1}$ $\times$ $km^{2}$ = $10^{-3}$ $m$ $h^{-1}$ $\times$ $10^{6}$ $m^{2}$ = $10^{3}$ $m^{3}$ $h^{-1}$

Volume should be devided by $10^{3}$ to make the unit $m^{3}$ $h^{-1}$

Mass = Volume $\times$ Density = ($10^{-3}$ $m^{3}$ $h^{-1}$) $\times$ $10^{3}$ $kg$ $m^{-3}$ = $kg$ $h^{-1}$

# Get hour from time string

```python
# Define a function to return hour from a byte string
def return_hour(datetimestring):
    # Decode bytes object to string
#     dtstr = datetimestring.decode()
    # datetimestring has this format: 'yyyy-mm-dd_hh:mm:'
    # hh is on string 11-12
#     hourstr = dtstr[11:13]
    hourstr = datetimestring[11:13]
    
    # Return -99 for an empty string
    if hourstr != '':
        return int(hourstr)
    else:
        return -99
```

```python
# Use list comprehension to loop over each track
# Then pass all time strings of a track to map, which then applies return_hour function to each string element
# This is essentially a fancy double for loop, but much cleaner
# np.asarray convert the list to a numpy array
hour_obs = np.asarray([list(map(return_hour, t)) for t in datetimesstring_obs])
hour_wrf1 = np.asarray([list(map(return_hour, t)) for t in datetimesstring_wrf1])
hour_wrf2 = np.asarray([list(map(return_hour, t)) for t in datetimesstring_wrf2])

# D = lambda g: g.decode()
# str_time = [list(map(D, t)) for t in datetimesstring_obs]
# print(str_time[0][0])
```

### A small function to format frequency axis to %

```python
def to_percent(y, position):
    # Ignore the passed in position. This has the effect of scaling the default
    # tick locations.
    s = str(100 * y)

    # The percent symbol needs escaping in latex
    if mpl.rcParams['text.usetex'] is True:
        return s + r'$\%$'
    else:
        return s + '%'
```
